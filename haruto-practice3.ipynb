```python
import pyautogui
import time
import keyboard

# ===== 設定 =====
STAMP_X = 1360   # スタンプをクリックするX座標
STAMP_Y = 289    # スタンプをクリックするY座標
INTERVAL = 0.1   # クリック間隔（秒）

# ★追加：入力欄（メッセージを打つ場所）の座標
TEXTBOX_X = 1000
TEXTBOX_Y = 950

# ★追加：送信ボタンの座標（Enterで送れない設定の人用。不要ならNone）
SEND_BUTTON = None  # 例: (1800, 950)
# =================

pyautogui.FAILSAFE = True
pyautogui.PAUSE = 0


def type_text_at_positions(positions, text, click_interval=0.05):
    """
    positions: [(x1,y1), (x2,y2), ...] のリスト
    text: 入力する文章（文字列）
    指定位置を順にクリックして、textを入力する
    """
    for (x, y) in positions:
        pyautogui.click(x, y)
        time.sleep(click_interval)
        pyautogui.write(text, interval=0.02)


def send_text_message(text, textbox_pos, send_button_pos=None):
    """
    textbox_pos: (x,y) 入力欄座標
    send_button_pos: (x,y) 送信ボタン座標（Enter送信できない場合に指定）
    """
    # 入力欄をクリック→文字入力
    type_text_at_positions([textbox_pos], text)

    # 送信（Enter or 送信ボタン）
    if send_button_pos is None:
        pyautogui.press('enter')
    else:
        pyautogui.click(send_button_pos[0], send_button_pos[1])


print("2秒後に開始（Shift + Q で終了）")
time.sleep(2)

# ★追加：スタンプ送信の直前に「はたらけ」を送る（1回だけ）
send_text_message("はたらけ", (TEXTBOX_X, TEXTBOX_Y), SEND_BUTTON)

# ここからスタンプ連打
while True:
    if keyboard.is_pressed('shift+q'):
        print("終了しました")
        break

    pyautogui.click(STAMP_X, STAMP_Y)
    time.sleep(INTERVAL)
```
